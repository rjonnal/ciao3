## How important is it for spots to be focused correctly on the SH sensor?

Idea:

1. Simulate pupil function (square?) at the lenslet.
2. Propagate beam various distances from lenslet, including focal length.
3. Simulate shot and read noise at the sensor.
4. Iterate and compute center of mass error.

```python
import numpy as np
from matplotlib import pyplot as plt

pitch_m = 300e-6
f_m = 14.2e-3
dxy = 1e-6 # simulation granularity
L = 800e-9 # wavelength

print(pitch_m/dxy)
assert pitch_m/dxy%1==0
N = int(pitch_m/dxy)


pfx,pfy = np.meshgrid(np.arange(-pitch_m/2,pitch_m/2,dxy),np.arange(-pitch_m/2,pitch_m/2,dxy))
focus_W_radians = (pfx**2+pfy**2)*2*np.pi/(2*f_m*L)

# add some tilt and tip
tip_angle_radians = 0.00
tilt_angle_radians = 0.00
tip_W_radians = pfx*np.tan(tip_angle_radians)*2*np.pi/L
tilt_W_radians = pfy*np.tan(tilt_angle_radians)*2*np.pi/L

W_radians = focus_W_radians + tip_W_radians + tilt_W_radians

pupil_function_complex = np.ones(pfx.shape)*np.exp(1j*W_radians)
dz = 14.2e-3

freq = np.fft.fftshift(np.fft.fftfreq(N,dxy))
freqx,freqy = np.meshgrid(freq,freq)
propagation = np.exp(1j*2*np.pi/L*dz)*np.exp(-1j*np.pi*L*dz*(freqx**2+freqy**2))

propagated = np.fft.ifftshift(np.fft.ifft2(np.fft.fft2(np.fft.fftshift(pupil_function_complex)) * propagation))
plt.imshow(np.abs(propagated))
plt.colorbar()
plt.show()
```
