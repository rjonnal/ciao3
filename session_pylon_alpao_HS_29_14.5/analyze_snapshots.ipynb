```python
import numpy as np
from matplotlib import pyplot as plt
import os,sys,glob,json
import pandas as pd

def load_dict(fn):
    with open(fn,'r') as fid:
        s = fid.read()
        d = json.loads(s)
    return d

spots_files = sorted(glob.glob(os.path.join('snapshots','*spots.npy')))
params_files = sorted(glob.glob(os.path.join('snapshots','*params.json')))
data_files = sorted(glob.glob(os.path.join('snapshots','*data.csv')))

k = 0
spots = np.load(spots_files[k])
params = load_dict(params_files[k])
data = pd.read_csv(data_files[k])
for k in params.keys():
    print(k,params[k])
```
